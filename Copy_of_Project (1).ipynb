### 1. How does temperature vary across different countries?

```python
#BAR PLOT
avg_temp=df.groupby("country")["temperature_celsius"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=avg_temp.values,y=avg_temp.index)
plt.title("Top 10 hottest countries")
plt.xlabel("Temperature (in celsius)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

### Inference
- Objective: The plot identifies the top 10 countries with the highest average temperatures.
- United Arab Emirates exhibits the highest mean temperature in the dataset.
- Trend: The ranking is dominated by countries with arid and desert climates.

```python
#BAR PLOT
avg_temp=df.groupby("country")["temperature_celsius"].mean().sort_values(ascending=False).tail(10)
plt.figure(figsize=(8,5))
sns.barplot(x=avg_temp.values,y=avg_temp.index)
plt.title("Least 10 hottest countries")
plt.xlabel("Temperature (in celsius)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

### Inference
- Objective: The plot identifies the 10 countries with the lowest average temperatures.
- Mongolia exhibits the lowest mean temperature in the dataset.
- Trend: The ranking is dominated by countries with high latitude, reflecting the expected climatic patterns of arctic and sub-arctic.

### 2. Which locations experiences extreme weather conditions?

```python
# BOX PLOT
plt.figure(figsize=(12,5))
top10_place = df['place'].value_counts().head(10).index
filtered_df = df[df['place'].isin(top10_place)]
sns.boxplot(data=filtered_df, x='place', y='temperature_celsius')
plt.title("Extreme Temperature Detection")
plt.show()
```

### Inference
- Extreme weather is most evident in Baghdad (most intense heat reaching ~ 50), N’djamena (high temperature with multiple outliers ~ frequent extreme heat spikes), and Kabul (experiences both cold and warm extremes).
- Vatican City shows large variability with lower temperatures dipping below 0°C → mixed conditions.
- Accra, Malabo, Suva have tight boxes → stable temperatures with minimal extremes.
- Asmara has moderate temperatures with low variation → relatively consistent climate.
- Tokyo & Bujumbura show moderate spread → some variation but fewer extreme outliers.

### 3. What is the distribution of humidity levels?

```python
#HISTOGRAM
plt.figure(figsize=(8,4))
sns.histplot(df['humidity'], bins=30, color='green', kde=True)
plt.title("Distribution of Humidity Levels")
plt.xlabel("Humidity")
plt.ylabel("Frequency")
plt.show()
```

### Inference
- The distribution is left-skewed (negatively skewed) → most values are on the higher humidity side.
- A large concentration of values lies between 60% and 90% humidity.
- There are very few low-humidity values (below 20%) → dry conditions are rare in the dataset.
- he peak (highest frequency) appears around 75–85% humidity → most common atmospheric condition.
- The long tail toward lower values indicates occasional dry conditions, but not frequent.

### 4. Is there a relationship between visibilty in miles and visibility in km?

### 6. Which weather factors are most strongly correlated with temperature?

```python
# HEATMAP
plt.figure(figsize=(10,4))
cols = ['temperature_celsius', 'humidity', 'cloud',
        'wind_mph', 'precip_mm', 'uv_index', 'feels_like_celsius']

sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Weather Correlation Matrix")
plt.show()
```

### Inference:
- Strong positive correlation between temperature and feels_like_celsius (~0.98)
- Moderate positive correlation:
 → temperature & uv_index (~0.49)
 → humidity & cloud (~0.51)
 → uv_index & feels_like_celsius (~0.44)
- Moderate negative correlation:
 → humidity & uv_index (~ -0.55)
 → temperature & humidity (~ -0.34)
- Weak or no correlation:
 → wind_mph with most variables
 → precip_mm with all variables
- Overall: temperature, humidity, and UV index show meaningful relationships,
- while wind speed and precipitation have minimal impact.
